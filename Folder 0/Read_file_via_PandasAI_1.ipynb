```python
# Import Libraries
import pandas as pd
from pandasai import SmartDatalake
from pandasai.llm import OpenAI
```

```python
#  Load OpenAI API key
api_key=open("/content/ChatGPTapiKey.txt","r").read().strip("\n")
```

```python
# Dataset Load
dataset = pd.read_csv("/content/gender_submission.csv")
```

```python
# Initialize the LLM Model
llm = OpenAI(api_token=api_key)
dl = SmartDatalake([dataset], config={"llm": llm})
```

```python
def askAboutFile(dl):
  while(1):
    print("Prompt(To end conversation type 'exit'): ")
    prompt=input()
    if(prompt.lower()=="exit"):
      print("Thank you for visiting!")
      break
    response=dl.chat(prompt)
    print("Response:\n",response)
    print()
```

```python
askAboutFile(dl)
```
